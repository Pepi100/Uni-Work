```python
#imports

import numpy as np
import matplotlib.pyplot as plt
import os
from skimage import io
```

```python
# Exercitii:
# a)
# loading

path = os.listdir('images/')
images = np.zeros((9,400,600))

index = 0
for img in path:
    images[index] = np.load('images/' +img)
    index+=1

    
    


```

```python
# b)
# sum of all pixels

print(np.sum(images))
```

```python
# c
# sum of all pixels for each image

sum_each =np.sum(images, axis=(1,2))
print(sum_each)
```

```python
# d
# maximum value index

print(np.argmax(sum_each))
```

```python
# e
# mean image

mean_image = np.mean(images, axis=0)

io.imshow(mean_image.astype(np.uint8)) 
io.show() 
```

```python
# f
# standard deviation
deviation =np.std(images)
print (deviation)
```
